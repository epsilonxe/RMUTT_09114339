```python
import matplotlib.pyplot as plt

plt.plot(5, 8, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plotting Example')
plt.show()
```

```python
import pandas as pd
import numpy as np

p_india = np.random.randint(5000, 8000, size=20)
p_china = np.random.randint(5000, 8000, size=20)
country_df = pd.DataFrame({'india':p_india, 'china':p_china})
```

```python
country_df.plot(kind='line')
plt.savefig('ex_country_plot.png')
```

```python
country_df['india'].plot(kind='hist')
plt.savefig('ex_country_hist.png')
```
